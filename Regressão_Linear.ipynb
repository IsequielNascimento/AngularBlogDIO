Carregando os dados de exemplo

```python
A = [ [1,2,58], [2,6,105], [3,8,88], [4,8,118], [5,12,117], [6,16,137],[7,20,157], [8,20,169],[9,22,149],[10,26,202]]

import numpy as np

arr = np.asmatrix(A)
X = np.concatenate((arr[:, 1], np.ones((arr.shape[0], 1))), axis=1)
Y = arr[:,2]

W0 = np.asmatrix([[4],[30]])

learning_rate = 0.001

W = W0
m = 0
```

Exibição da estimativa inicial

```python
import matplotlib.pyplot as plt

plt.plot(X[:,0], Y, 'o')
plt.xlabel('População')
plt.ylabel('Vendas')



# Plot the estimated function curve
x_vals = np.arange(X[:, 0].min(), X[:, 0].max(), 0.1)

y_vals = x_vals * W0[0,0] + W0[1,0]  # estimated function equation
plt.plot(x_vals, y_vals, "--")

plt.show()
```

Realização do Gradiente descendente
$W_{novo} = W_{antigo} - \eta \nabla J(W)$
onde <br>
$\nabla J(W) = \frac{1}{m} X^T (W_{antigo} X - Y )$ e <br>
 $\eta$ é a taxa de aprendizagem

```python
for k in range(1,10):
  W = W-X.transpose().dot(X.dot(W)-Y)/len(Y)*learning_rate;
m = m + k+1
```

Exibindo o resultado


```python
import matplotlib.pyplot as plt

plt.plot(X[:,0], Y, 'o')
plt.xlabel('População')
plt.ylabel('Vendas')


plt.title(str(m) +"ª iteração")
# Plot the estimated function curve
x_vals = np.arange(X[:, 0].min(), X[:, 0].max(), 0.1)

y_vals = x_vals * W[0,0] + W[1,0]  # estimated function equation
plt.plot(x_vals, y_vals, "--")

plt.show()
```
